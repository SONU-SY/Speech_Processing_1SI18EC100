# Sound 'a' from female

```python
from os.path import dirname, join as pjoin
from scipy.io import wavfile
import scipy.io
import numpy as np
import math 
from scipy.signal import hann
from scipy.fftpack import fft
import matplotlib.pyplot as plt
data_dir = pjoin(dirname(scipy.io.__file__), 'tests', 'data')
wav_fname = pjoin(data_dir, 'fa.wav')
samplerate, data = wavfile.read(wav_fname)
window = hann(160)
data_win=np.multiply(data[:160],window)
mags = abs(fft(data_win))
mags_dB=20*np.log10(mags)
N=len(data_win)
k=np.arange(N)
f=k*samplerate/N
plt.plot(f,mags_dB)
plt.ylabel("Magnitude (dB)")
plt.xlabel("Frequency Bin")
plt.title("Spectrum")
plt.grid()
```

Inference from 'a' from female:
*First Four Formants(Hz)(approximately):596,1012,1451,3012. 
*Manner of articulation:Vowel(voiced). 
*Place of  articulation:Low front tense. 
*Spacing between the bands is more compared to the respective sound of male.
